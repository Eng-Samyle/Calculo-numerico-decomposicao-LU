```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento

```python
# Calculo do faturamento total de cada loja
faturamento1 = loja['Preço'].sum()
faturamento2 = loja2['Preço'].sum()
faturamento3 = loja3['Preço'].sum()
faturamento4 = loja4['Preço'].sum()
print(f'O faturamento da loja 1 foi de R${faturamento1}, da loja 2 foi de R${faturamento2}, da loja 3 foi de R${faturamento3} e da loja 4 foi de R${faturamento4}.')
```

```python
def obter_faturamento_anual(loja, nome_loja):
    loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], format='%d/%m/%Y')
    loja['Ano'] = loja['Data da Compra'].dt.strftime('%Y')
    faturamento_por_ano = loja.groupby('Ano')['Preço'].sum().reset_index(name='Faturamento')
    faturamento_por_ano['Loja'] = nome_loja
    return faturamento_por_ano

# Obter os dados de faturamento de cada loja
faturamento_loja1 = obter_faturamento_anual(loja, "Loja 1")
faturamento_loja2 = obter_faturamento_anual(loja2, "Loja 2")
faturamento_loja3 = obter_faturamento_anual(loja3, "Loja 3")
faturamento_loja4 = obter_faturamento_anual(loja4, "Loja 4")

# Combinar todos os DataFrames em um único DataFrame
faturamento_total = pd.concat([faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4])

# Definir as cores para cada loja
cores = {'Loja 1': 'purple', 'Loja 2': 'blue', 'Loja 3': 'green', 'Loja 4': 'orange'}

# Criar o gráfico de linhas unificado
plt.figure(figsize=(8, 6))
plt.ylim(60000, 500000) # Manter o mesmo limite vertical para comparação

# Iterar sobre as lojas e plotar seus faturamentos
for nome_loja in faturamento_total['Loja'].unique():
    dados_loja = faturamento_total[faturamento_total['Loja'] == nome_loja]
    plt.plot(dados_loja['Ano'], dados_loja['Faturamento'], marker='o', linestyle='-', color=cores[nome_loja], label=nome_loja)

plt.xlabel('Ano')
plt.ylabel('Faturamento Anual')
plt.title('Faturamento Anual de Todas as Lojas')
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend() # Adicionar legenda para identificar as lojas
plt.tight_layout()
plt.show()
```

## Conclusão da avaliação do faturamento

### A loja que possuí o maior faturamento assim como a queda nas vendas menos agressiva é a loja 1, a loja 2 e 3 tem quadas muito bruscas, já a quarta também é brusca mas não tão agressiva como de outras, porém ela teve o menor faturamento entre todas as lojas, sendo assim ela está em pior situação nesse primeiro momento.

# 2. Vendas por Categoria

```python
def gerar_grafico_categorias(loja, nome_da_loja):
    eletronicos = 0
    moveis = 0
    brinquedos = 0
    eletrodomesticos = 0
    esporte_e_lazer = 0
    livros = 0
    utilidades_domesticas = 0
    instrumentos_musicais = 0

    for Categoria in loja['Categoria do Produto']:
        if Categoria == 'eletronicos':
            eletronicos += 1
        elif Categoria == 'moveis':
            moveis += 1
        elif Categoria == 'brinquedos':
            brinquedos += 1
        elif Categoria == 'eletrodomesticos':
            eletrodomesticos += 1
        elif Categoria == 'esporte e lazer':
            esporte_e_lazer += 1
        elif Categoria == 'livros':
            livros += 1
        elif Categoria == 'utilidades domesticas':
            utilidades_domesticas += 1
        elif Categoria == 'instrumentos musicais':
            instrumentos_musicais += 1

    categoriasQuant = [eletronicos, moveis, brinquedos, eletrodomesticos, esporte_e_lazer, livros, utilidades_domesticas, instrumentos_musicais]
    categorias = ['Eletronicos', 'Moveis', 'Brinquedos', 'Eletrodomesticos', 'Esporte\n e Lazer', 'Livros', 'Utilidades\n Domestricas', 'Instrumentos\n Musicais']

    plt.figure(figsize=(10, 4))  # aumenta um pouco o tamanho do gráfico
    barras = plt.bar(categorias, categoriasQuant, color='purple')
    plt.title(f'Vendas por Categoria da {nome_da_loja}')
    plt.ylabel('Quantidade Vendida')

    # Adiciona os valores acima das barras
    for i, barra in enumerate(barras):
        altura = barra.get_height()
        plt.text(barra.get_x() + barra.get_width() / 2, altura + 0.5, str(categoriasQuant[i]),
                 ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.ylim(0, max(categoriasQuant) + 100)  # adiciona espaço para os números acima das barras
    plt.tight_layout()
    plt.show()
    indice_mais_vendido = categoriasQuant.index(max(categoriasQuant))
    produto_mais_vendido = categorias[indice_mais_vendido]
    quantidade1 = eletronicos + moveis + brinquedos + eletrodomesticos + esporte_e_lazer + livros + utilidades_domesticas + instrumentos_musicais
    print(f' A quantidade de produtos vendidos na {nome_da_loja} foi {quantidade1}')
    print(f'📌 Produto mais vendido da {nome_da_loja}: {produto_mais_vendido} ({categoriasQuant[indice_mais_vendido]} vendas)\n')

gerar_grafico_categorias(loja, "Loja 1")
gerar_grafico_categorias(loja2, "Loja 2")
gerar_grafico_categorias(loja3, "Loja 3")
gerar_grafico_categorias(loja4, "Loja 4")
```

### Conclusão da analise de vendas por categorias

Note que a quantidade de produtos nas 3 primeiras lojas foi a mesma quantidade, na quarta foi apenas um produto a menos, porém a distribuição entre a venda das categorias é bem interressante, note pelos graficos que a categoria que mais vende produtos é a dos moveis em todas, porém a distribuição da quantidade é bem distribuida principalemente na loja 2, seguida da lojas 4. Compreender a distríbuição e qual é o produto mais vendido em cada loja é de extrema importância

# 4. Produtos Mais e Menos Vendidos


```python
from collections import Counter
def contagem_produtos(loja, nome_da_loja):
# Suponha que df seja seu DataFrame e 'Nome do Produto' a coluna
  lista_de_produtos_vendidos = loja['Produto'].tolist()

  contagem_de_produtos = Counter(lista_de_produtos_vendidos)
  # ... (o restante do código para encontrar o mais e menos vendido permanece o mesmo)
  mais_vendidos = contagem_de_produtos.most_common(1)
  if mais_vendidos:
      produto_mais_vendido = mais_vendidos[0][0]
      quantidade_mais_vendido = mais_vendidos[0][1]
      print(f"Na {nome_da_loja}\nO Produto mais vendido: {produto_mais_vendido} (Vendido {quantidade_mais_vendido} vezes)")
  else:
      print("Nenhum produto vendido encontrado.")
  if contagem_de_produtos:
      menor_contagem = min(contagem_de_produtos.values())
      menos_vendidos = [produto for produto, contagem in contagem_de_produtos.items() if contagem == menor_contagem]
      print(f"Produto(s) menos vendido(s): {', '.join(menos_vendidos)} (Vendido(s) {menor_contagem} vez(es))\n")
  else:
      print("Nenhum produto vendido encontrado para determinar o menos vendido.")
contagem_produtos(loja, "Loja 1")
contagem_produtos(loja2, "Loja 2")
contagem_produtos(loja3, "Loja 3")
contagem_produtos(loja4, "Loja 4")
```

Note que o produto mais vendido na loja 4 foi a cama box que em comparação com os outros itens é o mais caro
